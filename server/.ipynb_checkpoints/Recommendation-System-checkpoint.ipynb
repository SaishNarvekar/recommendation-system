```python
import numpy as np
import pandas as pd
```

```python
poi= pd.read_csv('persons.csv',index_col='Name',names = ['placeID','Name','Lat','Lng','Rating','Type'])
```

```python
reviews = pd.read_csv('reviews.csv',index_col='Name',names=['ID','Name','UserID','UserRating'])
```

```python
poi.drop(columns=['placeID','Lat','Lng'],axis=1,inplace=True)
```

```python
reviews.drop(columns=['ID'],axis=1,inplace=True)
```

```python
overall = poi.join(reviews,on='Name')
```

```python
overall_pvtable = overall.pivot_table(index='UserID',columns='Name',values='UserRating')
overall_pvtable
```

```python
BagaBeach = overall_pvtable['Colva Beach']
```

```python
similarToBaga = pd.DataFrame(overall_pvtable.corrwith(BagaBeach),columns=['Correlation'])
similarToBaga.dropna(inplace=True)
similarToBaga.sort_values(by='Correlation',ascending=False)
```

```python
final = similarToBaga.join(poi,on='Name').sort_values(by='Rating',ascending=False)
```
